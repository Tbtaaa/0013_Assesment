```python
# 1. Import The Packages to use in the project 
import os
import urllib.request
import pandas as pd
import geopandas as gpd

def ensure_data_folders():
    folders = ["data/raw", "data/clean"]
    
    for folder in folders:
        if not os.path.isdir(folder):
            os.makedirs(folder)
            print(f"Created folder: {folder}")
        else:
            print(f"Folder already exists: {folder}")

    print("--" * 30)

ensure_data_folders()

def download_if_missing(url, local_name):
    dest = os.path.join("data","raw", local_name)
    
    print(f" let me check: {local_name}")

    if os.path.isfile(dest):
        print(f"File already exists — skipping download.")
    else:
        print(f"Downloading from:\n     {url}")
        urllib.request.urlretrieve(url, dest)

        print(f"Download complete — saved")

    print("--" * 20)
    return dest 
```

```python
# ============================================================
# 4. CLEAN LISTINGS (EMILY & NADIA)
# ============================================================

# 4.1 Clean price column (remove $, commas, convert to numeric)
listings['price'] = (
    listings['price']
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)
listings['price'] = pd.to_numeric(listings['price'], errors='coerce')

# 4.2 Convert needed numeric columns
num_cols = [
    'availability_365',
    'estimated_occupancy_l365d',
    'estimated_revenue_l365d',
    'host_total_listings_count'
]
for col in num_cols:
    listings[col] = pd.to_numeric(listings[col], errors='coerce')

# 4.3 Normalise room_type
listings['room_type'] = listings['room_type'].str.lower().str.strip()

# 4.4 Professional host flag
listings['is_professional_host'] = listings['host_total_listings_count'] > 1

# 4.5 Keep only columns needed for Q3
cols_needed = [
    'id',
    'host_id',
    'latitude',
    'longitude',
    'price',
    'availability_365',
    'estimated_occupancy_l365d',
    'estimated_revenue_l365d',
    'host_total_listings_count',
    'room_type',
    'is_professional_host'
]
listings_q3 = listings[cols_needed].copy()

# 4.6 Filter to entire homes only
listings_entire = listings_q3[listings_q3['room_type'] == "entire home/apt"].copy()

# 4.7 Filter to full-time (90+ days available)
listings_fulltime = listings_entire[listings_entire['availability_365'] > 90].copy()

print("full-time entire homes:", listings_fulltime.shape)


# ============================================================
# 5. GEO-ENABLE LISTINGS + SPATIAL JOIN TO BOROUGHS
# ============================================================

import geopandas as gpd

listings_gdf = gpd.GeoDataFrame(
    listings_fulltime,
    geometry=gpd.points_from_xy(
        listings_fulltime.longitude,
        listings_fulltime.latitude
    ),
    crs="EPSG:4326"
)

listings_gdf = gpd.sjoin(
    listings_gdf,
    borough_clean,   # must contain ['borough_name','geometry']
    how='left',
    predicate='within'
)

# Clean borough names again after join
listings_gdf['borough_name'] = (
    listings_gdf['borough_name']
    .str.lower()
    .str.strip()
)

print("after spatial join:", listings_gdf.shape)


# ============================================================
# 6. MERGE RENT + COUNCIL TAX
# ============================================================

listings_gdf = listings_gdf.merge(
    rent[['borough_name', 'avg_annual_rent']],
    on='borough_name',
    how='left'
)

listings_gdf = listings_gdf.merge(
    council[['borough_name', 'council_tax']],
    on='borough_name',
    how='left'
)

# Drop listings missing rent or tax
listings_gdf = listings_gdf.dropna(
    subset=['avg_annual_rent', 'council_tax']
)

print("final Q3 dataset:", listings_gdf.shape)


# ============================================================
# 7. Q3 ANALYSIS MODEL
# ============================================================

# Keep only professional full-time entire home listings
listings_prof = listings_gdf[listings_gdf['is_professional_host']].copy()

print("professional full-time listings:", listings_prof.shape)

# Airbnb revenue proxy
listings_prof['airbnb_annual_revenue'] = (
    listings_prof['price'] * listings_prof['availability_365']
)

# Long-term annual rent
listings_prof['ltr_annual_revenue'] = listings_prof['avg_annual_rent']

# Current Airbnb advantage over long-term rent
listings_prof['net_advantage_current'] = (
    listings_prof['airbnb_annual_revenue']
    - listings_prof['ltr_annual_revenue']
)


# ============================================================
# 8. POLICY MODEL — Band D * 2 (or 3)
# ============================================================

multiplier = 10  # 2x Band D council tax per listing

listings_prof['policy_charge'] = listings_prof['council_tax'] * multiplier

listings_prof['net_advantage_after'] = (
    listings_prof['airbnb_annual_revenue']
    - listings_prof['policy_charge']
    - listings_prof['ltr_annual_revenue']
)


# ============================================================
# 9. FLAG AFFECTED LISTINGS
# ============================================================

listings_prof['affected'] = (
    (listings_prof['net_advantage_after'] < 0) |
    (
        (listings_prof['net_advantage_current'] > 0) &
        (listings_prof['net_advantage_after']
         < 0.5 * listings_prof['net_advantage_current'])
    )
)

n_total = listings_prof.shape[0]
n_affected = int(listings_prof['affected'].sum())
pct_affected = n_affected / n_total * 100

print(
    f"out of {n_total:,} professional full-time listings, "
    f"{n_affected:,} ({pct_affected:.1f}%) are affected under this policy."
)


# ============================================================
# 10. BOROUGH SUMMARY
# ============================================================

borough_summary = (
    listings_prof
    .groupby('borough_name')
    .agg(
        n_listings=('id', 'count'),
        n_affected=('affected', 'sum'),
        mean_airbnb_revenue=('airbnb_annual_revenue', 'mean'),
        mean_ltr_revenue=('ltr_annual_revenue', 'mean'),
        mean_council_tax=('council_tax', 'mean')
    )
    .reset_index()
)

borough_summary['pct_affected'] = (
    borough_summary['n_affected'] / borough_summary['n_listings'] * 100
)

print(
    borough_summary
    .sort_values('pct_affected', ascending=False)
    .head(10)
)


# ============================================================
# 11. SCATTER PLOT (profit before vs after policy)
# ============================================================

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

plt.scatter(
    listings_prof['net_advantage_current'],
    listings_prof['net_advantage_after'],
    s=5, alpha=0.4, label='unaffected'
)

plt.scatter(
    listings_prof.loc[listings_prof['affected'], 'net_advantage_current'],
    listings_prof.loc[listings_prof['affected'], 'net_advantage_after'],
    s=8, alpha=0.7, label='affected'
)

plt.axhline(0, linestyle='--')
plt.axvline(0, linestyle='--')

plt.xlabel('net advantage current (£/year)')
plt.ylabel('net advantage after policy (£/year)')
plt.title('impact of council tax surcharge on professional full-time airbnb listings')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

multipliers = [2, 5, 10, 15, 20]

rows = []

# build borough–multiplier–%affected table
for m in multipliers:
    tmp = listings_prof.copy()

    # net advantage after this multiplier
    tmp['net_advantage_after_m'] = (
        tmp['airbnb_annual_revenue']
        - (tmp['council_tax'] * m)
        - tmp['ltr_annual_revenue']
    )

    # affected flag for this multiplier
    tmp['affected_m'] = (
        (tmp['net_advantage_after_m'] < 0) |
        (
            (tmp['net_advantage_current'] > 0) &
            (tmp['net_advantage_after_m'] < 0.5 * tmp['net_advantage_current'])
        )
    )

    # borough-level % affected
    borough_m = (
        tmp.groupby('borough_name')['affected_m']
        .value_counts(normalize=False)
        .unstack(fill_value=0)
        .reset_index()
    )

    borough_m['total'] = borough_m[False] + borough_m[True]
    borough_m['pct_affected'] = 100 * borough_m[True] / borough_m['total']
    borough_m['multiplier'] = m

    rows.append(borough_m[['borough_name', 'multiplier', 'pct_affected']])

borough_mult = pd.concat(rows, ignore_index=True)

# choose an order for boroughs (e.g. alphabetical)
borough_order = sorted(borough_mult['borough_name'].unique())
x = np.arange(len(borough_order))

plt.figure(figsize=(10, 6))

for m in multipliers:
    sub = borough_mult[borough_mult['multiplier'] == m].copy()
    # align to the fixed borough order
    sub = sub.set_index('borough_name').reindex(borough_order)
    plt.plot(
        x,
        sub['pct_affected'],
        marker='o',
        label=f'{m}×'
    )

plt.xticks(x, [b.title() for b in borough_order], rotation=90)
plt.ylabel('% of professional full-time listings affected')
plt.xlabel('Borough')
plt.title('Impact of increasing council tax on professional Airbnb listings\nby borough and multiplier')
plt.grid(axis='y', alpha=0.3)
plt.legend(title='Multiplier (× Band D)', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
